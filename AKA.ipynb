```python
import matplotlib.pyplot as plt

# Jumlah data
sizes = [20, 100, 1000]

# Average case (data tengah)
iter_avg = [1.2750, 3.3920, 5.2390]
rec_avg  = [1.4660, 5.1930, 10.8930]

# Worst case hanya di n = 1000
iter_worst = [None, None, 6.0860]   # contoh worst case found / not found
rec_worst  = [None, None, 11.1160]

plt.figure(figsize=(9, 6))

# Average case lines
plt.plot(sizes, iter_avg, marker='o', label='Iteratif (Average Case)')
plt.plot(sizes, rec_avg, marker='o', label='Rekursif (Average Case)')

# Worst case points (dipisah, bukan garis penuh)
plt.plot(sizes, iter_worst, marker='X', linestyle='None',
         markersize=10, label='Iteratif (Worst Case, n=1000)')
plt.plot(sizes, rec_worst, marker='X', linestyle='None',
         markersize=10, label='Rekursif (Worst Case, n=1000)')

plt.xlabel('Jumlah Data (n)')
plt.ylabel('Waktu Eksekusi (ms)')
plt.title('Analisis Waktu Eksekusi Binary Search\nIteratif vs Rekursif')

plt.legend()
plt.grid(True)

plt.show()
```
