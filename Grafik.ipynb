```python
import matplotlib.pyplot as plt

# Semua ukuran data
sizes = [20, 100, 1000]

# Average case (nilai tengah SEMUA n)
iter_avg = [0.0000, 0.0000,  0.5930]
rec_avg  = [1.5060, 0.5260, 0.6880]

# Worst case hanya dianalisis di n = 1000
iter_worst_found = [6.9640]
rec_worst_found  = [4.7450  ]

iter_worst_nf = [1.6790 ]    # contoh jika kamu punya datanya
rec_worst_nf  = [2.1830]

plt.figure(figsize=(9, 6))

# Average case lines
plt.plot(sizes, iter_avg, marker='o',
         label='Iteratif (Average Case)')
plt.plot(sizes, rec_avg, marker='o',
         label='Rekursif (Average Case)')

# Worst case points at n=1000
plt.scatter([1000], iter_worst_found,
            marker='X', s=120,
            label='Iteratif (Worst Case - Found)')
plt.scatter([1000], rec_worst_found,
            marker='X', s=120,
            label='Rekursif (Worst Case - Found)')

plt.scatter([1000], iter_worst_nf,
            marker='D', s=100,
            label='Iteratif (Worst Case - Not Found)')
plt.scatter([1000], rec_worst_nf,
            marker='D', s=100,
            label='Rekursif (Worst Case - Not Found)')

plt.xlabel('Jumlah Data (n)')
plt.ylabel('Waktu Eksekusi (ms)')
plt.title('Perbandingan Waktu Eksekusi Binary Search\nIteratif vs Rekursif')

plt.grid(True)
plt.legend()
plt.show()
```
